```python
import numpy as np

array1=np.random.randint(0,100,(4,3))
array2=np.random.randint(0,100,(4,3))
array4=np.matmul(array1,array2.T)
print('array1: ',array1)
print('array2: ',array2)
print('array4: ',array4)
```
