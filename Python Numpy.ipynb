```python
import numpy as np
avg=np.mean(dataset[:,1:4],axis=1)
print(avg.reshape(-1,1))
```

```python
def findGrade(mark):
    if(mark>=0 and mark<35):
        grade='F'
    elif(mark>=35 and mark<45):
        grade='S'
    elif(mark>=45 and mark<65):
        grade='C'
    elif(mark>=65 and mark<75):
        grade='B'
    elif(mark>=75 and mark<100):
        grade='A'
    return grade
```

```python
mathsGrade=np.array([findGrade(i) for i in dataset[:,1]])
print(mathsGrade.reshape(-1,1))
```

```python
phyGrade=np.array([findGrade(i) for i in dataset[:,2]])
print(phyGrade.reshape(-1,1))
```

```python
cheGrade=np.array([findGrade(i) for i in dataset[:,3]])
print(cheGrade.reshape(-1,1))
```

```python
print(dataset.shape)
print(mathsGrade.shape)

print(dataset.ndim)
print(mathsGrade.ndim)
```
